This is Alex Myrick's notebook that deals with the median age of death in each neighborhood in Allegheny County. It's important to note that we are not tracking the location of the death, but rather, the registered residential area of individuals at their time of death. Obviously, this is a pretty important statistic; ultimately, the median age of death indicates the overall quality of a neighborhood, encompassing its air quality, wealth, housing effectiveness, health facilities, public sanitation, and much more that would be extremely difficult to relate into a single factor if not for this data set. 


```python
import pandas as pd
df = pd.read_csv("/home/jupyter-acm272/BigIdeasFinalProjectFolder/nhood.csv")
ddf = df.nlargest(5, "TOTAL MD AGE AT DEATH")
ddf.plot(ddf.columns[0], ddf.columns[6], kind="bar", rot=0, figsize=(10,10), title = "Highest MD Age of Death Neighborhoods vs MD Age of Death", ylabel = "MD Age of Death", legend=False)





```

Honestly, the median age of death is quite similar for all of these neighborhoods, but I don't think it's a coincidence that Squirrel Hill South has both the greatest number of deaths and the highest median age of death; if it were not for the lurking variables that could be present here, I'd immediately label this as a Law of Large Numbers-esque result, but I cannot in good conscience do so because we're actually looking to disprove this idea here by identifying underlying health issues in neighborhoods. For reference, the mean median death age is 74.8483 years. If we're going to look at how much these data actually mean, however, we need to see whether they're actually relevant in terms of overall deaths. Here's a graph of the 10 neighborhoods with the highest median death age with regards to their total number of deaths:

```python
df = pd.read_csv("/home/jupyter-acm272/BigIdeasFinalProjectFolder/nhood.csv")
fdd = df.nlargest(10, "TOTALdeaths*")
fdd.plot(ddf.columns[0], ddf.columns[5], kind="bar", rot=0, figsize=(20,20), title = "Highest Death Count Neighborhoods vs Total Deaths", ylabel = "Total Deaths", legend=False)
```
